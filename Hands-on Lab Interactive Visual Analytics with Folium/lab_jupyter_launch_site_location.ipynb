```python
import folium

# Lista dos launch sites com suas coordenadas
launch_sites = {
    'CCAFS LC-40': [28.562302, -80.577356],
    'CCAFS SLC-40': [28.563197, -80.576820],
    'KSC LC-39A': [28.573255, -80.646895],
    'VAFB SLC-4E': [34.632834, -120.610745]
}

# Coordenadas médias para centralizar o mapa
avg_lat = sum(coords[0] for coords in launch_sites.values()) / len(launch_sites)
avg_lon = sum(coords[1] for coords in launch_sites.values()) / len(launch_sites)
nasa_coordinate = [avg_lat, avg_lon]

# Inicializar o mapa centralizado na média das coordenadas
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Adicionar círculos para cada site
for site, coords in launch_sites.items():
    folium.Circle(
        location=coords,
        radius=500,  # metros
        color='blue',
        fill=True,
        fill_color='cyan',
        fill_opacity=0.5,
        popup=site
    ).add_to(site_map)

# Exibir o mapa
site_map
```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python
import folium
from folium.plugins import MarkerCluster

# Função para definir cor do marcador baseado no resultado do lançamento
def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
# Aplicar a cor dos marcadores
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Inicializar o mapa (exemplo usando coordenadas médias)
avg_lat = spacex_df['Lat'].mean()
avg_lon = spacex_df['Long'].mean()
site_map = folium.Map(location=[avg_lat, avg_lon], zoom_start=5)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
# Criar MarkerCluster
marker_cluster = MarkerCluster()
site_map.add_child(marker_cluster)

# Adicionar marcadores ao cluster
for index, row in spacex_df.iterrows():
    marker = folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=f"Launch Site: {row['Launch Site']}\nClass: {row['class']}",
        icon=folium.Icon(color=row['marker_color'])
    )
    marker_cluster.add_child(marker)

# Exibir mapa
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon

# Coordenadas do local de lançamento (ex: CCAFS LC-40)
launch_site = [28.562302, -80.577356]

# Coordenadas do ponto mais próximo da costa (exemplo manual)
coastline_point = [28.5630, -80.5670]

# Calcular a distância
distance = calculate_distance(launch_site[0], launch_site[1], coastline_point[0], coastline_point[1])

# Criar o mapa
site_map = folium.Map(location=launch_site, zoom_start=13)

# Adicionar marcador do local de lançamento
folium.Marker(
    location=launch_site,
    popup='CCAFS LC-40',
    icon=folium.Icon(color='blue')
).add_to(site_map)

# Adicionar marcador do ponto da costa
folium.Marker(
    location=coastline_point,
    popup='Ponto da Costa',
    icon=folium.Icon(color='green')
).add_to(site_map)

# Adicionar marcador com a distância
distance_marker = folium.Marker(
    location=coastline_point,
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:.2f} KM</b></div>'.format(distance),
    )
)
distance_marker.add_to(site_map)

# Adicionar linha conectando os pontos
folium.PolyLine(locations=[launch_site, coastline_point], color='gray', weight=2.5, opacity=0.7).add_to(site_map)

# Exibir o mapa
from IPython.display import display
display(site_map)
```
